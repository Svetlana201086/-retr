```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.stattools import adfuller
import pmdarima as pm

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

plt.style.use('seaborn-whitegrid')
```

```python
from statsmodels.datasets import get_rdataset
```

```python
data = get_rdataset('AirPassengers').data
ts = pd.Series(data['value'].values,
               index=pd.date_range('1949-01-01', periods=len(data), freq='M'),
               name='passengers')

ts.plot(figsize=(12,4), title='AirPassengers—1950–1960')
plt.show()
```

```python
adf_result = adfuller(ts)
print(f"ADF Statistic: {adf_result[0]:.4f}, p-value: {adf_result[1]:.4f}")
# p-value >> 0.05 → нестационарный ряд
```

```python
train, test = ts[:'1958-12'], ts['1959-01':]
print(f"train: {len(train)} точек, test: {len(test)} точек")
```

```python
# seasonal m=12
model = pm.auto_arima(train,
                      start_p=0, start_q=0,
                      max_p=3, max_q=3,
                      seasonal=True, m=12,
                      start_P=0, start_Q=0,
                      max_P=2, max_Q=2,
                      d=None, D=1,          
                      trace=True,
                      stepwise=True,
                      error_action='ignore',
                      suppress_warnings=True,
                      information_criterion='aic')

print(model.summary())
```
