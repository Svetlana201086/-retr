```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('data/Series/monthly-sales-of-company-x-jan-6.csv', 
                   index_col=0, parse_dates=[0])
series = data.iloc[:, 0]
```

```python
plt.figure(figsize=(12, 4))
plt.plot(series)
plt.title('Месячные продажи компании X')
plt.show()
```

```python
train_size = int(len(series) * 0.8)
train, test = series[:train_size], series[train_size:]
```

```python
import itertools

p = d = q = range(0, 3)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in pdq]

best_aic = np.inf
best_pdq = None
best_seasonal_pdq = None

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            model = ARIMA(train, order=param, seasonal_order=param_seasonal)
            results = model.fit()
            aic = results.aic
            if aic < best_aic:
                best_aic = aic
                best_pdq = param
                best_seasonal_pdq = param_seasonal
        except:
            continue

print(f'Лучшие параметры: {best_pdq}, {best_seasonal_pdq}')
```

```python
model = ARIMA(train, order=best_pdq, seasonal_order=best_seasonal_pdq)
results = model.fit()
```

```python
residuals = results.resid
plt.figure(figsize=(12, 4))
plt.plot(residuals)
plt.title('Остатки')
plt.show()
```

```python
# QQ-plot остатков
import statsmodels.api as sm
qqplot = sm.qqplot(residuals, line='s')
plt.title('QQ-plot остатков')
plt.show()
```

```python
# Создание лагированных признаков
def create_lagged_features(series, lag=12):
    X = pd.DataFrame({
        f'lag_{i}': series.shift(i) for i in range(1, lag + 1)
    })
    y = series
    return X.dropna(), y[X.index]
```

```python

X_train, y_train = create_lagged_features(train, lag=12)
X_test, y_test = create_lagged_features(test, lag=12)
```

```python
# Обучение модели Random Forest
from sklearn.ensemble import RandomForestRegressor
model_rf = RandomForestRegressor()
model_rf.fit(X_train, y_train)
```

```python
# Предсказание
predictions_rf = model_rf.predict(X_test)
mse_rf = mean_squared_error(y_test, predictions_rf)
print(f'MSE Random Forest: {mse_rf}')
```
